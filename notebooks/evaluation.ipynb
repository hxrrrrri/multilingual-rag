# RAG Pipeline Evaluation

This notebook evaluates the RAG pipeline using RAGAS metrics.

```python
import requests
import json

BASE_URL = 'http://localhost:8000/api/v1'

test_queries = [
    'What are the main eligibility criteria?',
    'What is the deadline mentioned?',
    'Summarise the key findings.',
]

results = []
for q in test_queries:
    resp = requests.post(f'{BASE_URL}/query', json={'query': q})
    results.append(resp.json())
    print(f'Q: {q}')
    print(f'A: {resp.json()["answer"][:200]}')
    print(f'Faithfulness: {resp.json()["faithfulness_score"]:.3f}')
    print()
```
